```python
import numpy as np
import math
def gen_kernel(size):
    
    kernel =  np.fromfunction(lambda x, y: (1/(2*math.pi*1**2)) * math.e ** ((-1*((x-(size-1)/2)**2+(y-(size-1)/2)**2))/(2*1**2)), (size, size))
    return kernel / np.sum(kernel)
```

```python
def gauss_Filter(mg,kernel_size):
    kernel = gen_kernel(kernel_size)
    h = int((kernel_size-1)/2)
    gaussian = np.zeros((mg.shape[0]-2*h, mg.shape[1]-2*h))
    for x in range(h, mg.shape[0]-h):
        for y in range(h, mg.shape[1]-h):
            gaussian[x-h][y-h] = np.sum(np.multiply(mg[x-h:x+h+1, y-h:y+h+1], kernel))
    return gaussian
```

```python
mg= np.random.rand(12,12)
print(mg)
```

```python
gauss_Filter(mg,3)
```
